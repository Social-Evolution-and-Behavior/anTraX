```python
## behavioral measures
import itertools
import numpy as np
import scipy

idx = pd.IndexSlice
root = '/Volumes/Jonathan/'
wd = root + 'analysis'
datafile = join(wd,'antdata.h5')

# the binsize
bs = 24*3600*10
drad = 0.005
intrad = 0.002
M = []

spatial_bs = 0.005

def spatial_entropy(x):
    
    
    h = np.histogram2d(a,b,20) 
    ent = scipy.stats.entropy(h.flatten())
    
    return x
    
    
def meanseqdur(x):
    
    # number of sequences
    nseq = (x.astype('float64').diff()==1).sum()
    
    # seq pr
    sseq = x.sum()
    
    # mean seq dur
    mseq = sseq/nseq
    
    return mseq
    

    
IG = {}

for T in range(3, 13):
    
    Tstr = 'T' + str(T)
    
    print('Working on ' + Tstr)
    print('  ...reading')
    ad = pd.read_hdf(join(wd,'antdata_' + Tstr + '.h5'),'ad')
    print('  ...computing')
    groups = np.floor(ad.index.values / bs)
    
    meas = []
    keys = []
                      
    # activity level (total distance traveled)
    df = ad.loc[:,idx[:,'v']].groupby(groups).sum().droplevel(axis=1,level=1)
    meas.append(df)
    keys.append('activity')
    
    # nesting (prop time in the nest)
    df = 1 - ad.loc[:,idx[:,'outside']].groupby(groups).mean().droplevel(axis=1,level=1)                 
    meas.append(df)
    keys.append('nesting')
    
    # nesting (prop time in the nest)
    df = ad.loc[:,idx[:,'outside']].droplevel(axis=1,level=1) 
    df = ~df
    df = df.groupby(groups).apply(meanseqdur)
    meas.append(df)
    keys.append('mean_nesting_duration')
    
    # nesting (prop time in the nest)
    df = ad.loc[:,idx[:,'outside']].droplevel(axis=1,level=1)
    df = df.groupby(groups).apply(meanseqdur)
    meas.append(df)
    keys.append('mean_exploration_duration')
    
    # aux 
    out = ad.loc[:,idx[:,'outside']].droplevel(axis=1,level=1).apply(lambda x: x.where(x, np.nan))
    
    # tendency to explore far when outside
    df = ad.loc[:,idx[:,'dnest']].droplevel(axis=1,level=1).copy() * out
    df = df.groupby(groups).mean()
    meas.append(df)
    keys.append('mean_exploration_rad')

    # velocity outside 
    df = ad.loc[:,idx[:,'v']].droplevel(axis=1,level=1).copy() * out
    df = df.groupby(groups).mean()
    meas.append(df)
    keys.append('mean_velocity')
    
    # acceleration
    df = ad.loc[:,idx[:,'acceleration']].droplevel(axis=1,level=1).copy() * out
    df = df.groupby(groups).mean()
    meas.append(df)
    keys.append('mean_acc')
    
    # acceleration
    df = ad.loc[:,idx[:,'normal_acceleration']].droplevel(axis=1,level=1).copy() * out
    df = df.groupby(groups).mean()
    meas.append(df)
    keys.append('normal_acc_outside')
    
    # curvature
    df = ad.loc[:,idx[:,'curvature']].droplevel(axis=1,level=1).copy() * out
    df = df.groupby(groups).mean()
    meas.append(df)
    keys.append('curvature')
    

    # interaction rate
    df = ad.loc[:,idx[:,'interacting']].droplevel(axis=1,level=1) 
    df = df.groupby(groups).mean()
    meas.append(df)
    keys.append('interaction_rate')
    
    # interaction duration
    df = ad.loc[:,idx[:,'interacting']].droplevel(axis=1,level=1).groupby(groups).apply(meanseqdur)
    meas.append(df)
    keys.append('mean_interaction_duration')
    
    # stop rate
    df = ad.loc[:,idx[:,'stop']].droplevel(axis=1,level=1) * out
    df = df.groupby(groups).mean()
    meas.append(df)
    keys.append('stop_rate')
    
    # stop duration
    df = ad.loc[:,idx[:,'stop']].droplevel(axis=1,level=1) * out
    df = df.groupby(groups).apply(meanseqdur)
    meas.append(df)
    keys.append('mean_stop_duration')
    
    
    # spatial entropy when outside the nest
    #x = ad.loc[:,idx[:,'x']].droplevel(axis=1,level=1) * out
    #y = ad.loc[:,idx[:,'y']].droplevel(axis=1,level=1) * out
    
    #df['new_col'] = list(zip(df.lat, df.long))

    # tendency to follow trails?
    
    # interaction graph
    IG[Tstr] = np.zeros((16, 16))
    for i, ai in enumerate(ex.antlist):
        for j, aj in enumerate(ex.antlist):
            dx = ad.loc[:,idx[ai,'x']] - ad.loc[:,idx[aj, 'x']] 
            dy = ad.loc[:,idx[ai,'y']] - ad.loc[:,idx[aj, 'y']]
            II = (np.sqrt(dx*dx + dy*dy) < 0.002) * out[ai] * out[aj]
            IG[Tstr][i, j] = II.mean()
        
    
    M.append(pd.concat(meas,axis=1,keys=keys))
    

M = pd.concat(M,axis=1,keys=['T'+str(x) for x in range(3,13)])
M.to_hdf(join(wd,'FeatureTable.h5'),'M')    
```

```python
features = set([m[1] for m in M.columns])

def shuffle(df):     
    df = df.copy()
    df = df.apply(np.random.permutation, axis=1, raw=True)
    return df

def consistency(df):
    r = df.rank(axis=1)
    n = r.shape[1]
    c = np.log(((r<n/2).sum()+1) / ((r>=n/2).sum()+1))
    return c.values


C = {}
C1 = {}
C2 = {}
SC =  {}


half = np.floor(M.shape[0]/2)

for f in features:
    
    MM = M.loc[:,idx[:,f,:]]
    C[f] = consistency(MM)
    
    
    MM1 = M.loc[:half,idx[:,f,:]]
    MM2 = M.loc[half:,idx[:,f,:]]
    
    C1[f] = consistency(MM1)
    C2[f] = consistency(MM2)
    
    sc = []
    for _ in range(100):
        sc.append(consistency(shuffle(MM)))

    SC[f] = np.concatenate(sc)
    
    

```

```python
# antenation power


def angle(x1,y1,x2,y2):
    dx = x2 - x1
    dy = y2 - y1
    return np.arctan2(dx, dy) 

def to_angle(x, per=2*np.pi):    
    return (x + per/2) % per - per/2;

for ant in ad.antlist:

    antdata = ad.data.loc[:,idx[ant,:]].droplevel(axis=1,level=0)
    box_head_angle = angle(antdata['Head_x'],antdata['Head_y'],antdata['Neck_x'],antdata['Neck_y'])
    box_r_ant_angle = angle(antdata['R_ant_tip_x'],antdata['R_ant_tip_y'],antdata['R_ant_root_x'],antdata['R_ant_root_y'])
    box_l_ant_angle = angle(antdata['L_ant_tip_x'],antdata['L_ant_tip_y'],antdata['L_ant_root_x'],antdata['L_ant_root_y'])
    
    head_r_ant_angle = to_angle(box_r_ant_angle - box_head_angle, per=np.pi)
    head_l_ant_angle = to_angle(box_l_ant_angle - box_head_angle, per=np.pi)
    ad.data[ant, 'r_ant_angle'] =  head_r_ant_angle
    ad.data[ant, 'l_ant_angle'] =  head_l_ant_angle
    

for ant in ad.antlist:
    ad.data[ant, 'antpower'] = (ad.data[ant, 'r_ant_angle'].diff()**2 + ad.data[ant, 'l_ant_angle'].diff()**2).rolling(51, center=True).mean()
```
