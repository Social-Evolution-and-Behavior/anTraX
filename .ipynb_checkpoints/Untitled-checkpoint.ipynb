```python
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
```

```python
THETA = 40
ETHA = 1
ALPHA = 0.1

T_BASELINE = 26

dt = 0.01
KR = 0.01


def sigmoid(theta, etha, x):
    
    return 1/(1+np.exp(-(x-theta)/etha))


class Ant:
    
    def __init__(self, theta=THETA, etha=ETHA, alpha=ALPHA):
        
        self.theta = theta
        self.etha = etha
        self.alpha = alpha
        
        self.state = False
        
    def step(self, T, S):
        
        T1 = T + self.alpha*S
        
        kp = sigmoid(self.theta, self.etha, T1)
        kr = 1 - sigmoid(self.theta, self.etha, T1)
        
        
        if self.state:
            
            self.state = rand() > kr*dt
            
        else:
            
            self.state = rand() < kp*dt
        
      
        
class Colony:
    
    def __init__(self, N=100, theta=THETA, etha=ETHA, alpha=ALPHA):
    
        self.N = N
        self.etha = etha
        self.theta = theta
        self.alpha = alpha
        
        self.reset()
        
    def reset(self):
        
        self.ants = [Ant(theta=self.theta, etha=self.etha, alpha=self.alpha) for _ in range(self.N)]
        self.S = self.calc_S()
        
    
    def step(self, T=T_BASELINE):
        
        for ant in self.ants:
            ant.step(T, self.S)
            
        self.S = self.calc_S()
        
    def state(self):
        
        return float(sum([a.state for a in self.ants]))/self.N
            
    def calc_S(self):
        A = np.array([a.state for a in self.ants]).astype('float')
        #S = -np.sum(~A)
        S = -40*np.mean(A-0.5)
        
        return S
    
    def simulate(self, TPERT, duration=120):
        
        self.reset()
        tt = np.arange(0, duration, dt)
        yy = np.full_like(tt, np.nan, dtype=np.double)
        
        for i, t in enumerate(tt):
            yy[i] = self.state()
            self.step(T=TPERT)
            
        return yy
    
    def find_threshold(self, K=100, TS=range(T_BASELINE,50), duration=120):
        
        yy = []
        for T in TS:
            #print(T)
            yyt = []
            for k in range(K):
                yyk = self.simulate(T, duration=duration)
                yyt.append(yyk)
            yy.append(yyt)
            
        return yy
            
    
```

```python
NS = [10,30,50,70,90,120,160,200]

yy_sym=[]
for N in NS:
    print(N)
    c = Colony(N, alpha=0.1, theta=32, etha=5)
    yy_sym.append(c.find_threshold(K=10))
```
